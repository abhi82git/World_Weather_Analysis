```python
# Import the dependencies.
import pandas as pd
import requests
import gmaps
# Import the API key.
from config import g_key
from IPython.display import display
```

```python
# Assign locations to array of latitude and longitude
Locations = city_data_df[["Latitude","Longitude"]]

# Get the max temp
max_temp = city_data_df["Max Temp"]

# Initialize an empty list for temps > 0
temps = []

for temp in max_temp:
    temps.append(max(temp,0))

# Assign the figure variable
fig = gmaps.figure(center=(30.0, 31.0), zoom_level=1.5)

# Assign the heatmap variable
heat_layer = gmaps.heatmap_layer(Locations,weights=temps,dissipating=False, max_intensity=300, point_radius=4)

# add the heatmap layer
fig.add_layer(heat_layer)

# Call the figure to plot the data
fig
```

```python
# Ask the user to input the minimum temperature
min_temp = float(input("What is the minimum temperature that you would like for your trip? "))

# Ask the user to input the maximum temperature
max_temp = float(input("What is the maximum temperature that you would like for your trip? "))

preferred_cities_df  = city_data_df.loc[(city_data_df["Max Temp"] <= max_temp) & (city_data_df["Max Temp"] >= min_temp)]
preferred_cities_df
```

```python
preferred_cities_df.notnull().sum()
```

```python
# Create DataFrame called hotel_df to store hotel names along with city, country, max temp, and coordinates.

hotel_df = preferred_cities_df[["City","Country", "Max Temp", "Latitude","Longitude"]].copy()
hotel_df["Hotel_Name"] = ""
hotel_df
```

```python
params = {
    "radius": 5000,
    "types": "lodging",
    "key": g_key,
    "location": ""
}

for index, row in hotel_df.iterrows():
    lat = row["Latitude"]
    lon = row["Longitude"]
    
    params["location"] = f"{lat},{lon}"
    
    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
    
    hotels = requests.get(base_url,params=params).json()
    
    try:
        hotel_df.loc[index, "Hotel Name"] = hotels["results"][0]["name"]
    except(IndexError):
        print("Hotel not found... skipping.")
```

```python
hotel_df.head(100)
```

```python
info_box_template = """
<dl>
<dt>Hotel Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
<dt>Max Temp</dt><dd>{Max Temp} °F</dd>
</dl>
"""
```

```python
# Store the DataFrame Row.
hotel_info = [info_box_template.format(**row) for index, row in hotel_df.iterrows()]

# Assign locations to array of latitude and longitude
Locations = hotel_df[["Latitude","Longitude"]]

# Get the cloudiness
max_temp = hotel_df["Max Temp"]

# Assign the figure variable
fig = gmaps.figure(center=(30.0, 31.0), zoom_level=1.5)

# Assign the heatmap variable
heat_layer = gmaps.heatmap_layer(Locations,weights=max_temp,dissipating=False, max_intensity=300, point_radius=4)

# Assign the marker layer
marker_layer = gmaps.marker_layer(Locations, info_box_content=hotel_info)

# add the heatmap layer
fig.add_layer(heat_layer)

# add the marker layer
fig.add_layer(marker_layer)

# Call the figure to plot the data
fig
```
